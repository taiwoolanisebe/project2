## Analysis Overview
In this analysis, I decided to pay attention to the composition and distribution of real estate properties in the districts of Tashkent, the capital of Uzbekistan, and how these information relate with the price of real estate properties in the State.

## Dataset Overview
This dataset consists of numerous attributes of more than 7,000 real estate properties that are up for sale in 12 districts in Tashkent, Uzbekistan. The various attributes in this data set include the address of properties, the districts of the state in which properties are located, the number of rooms each property comes with, the size of properties in square metres, the level or floor in which these properties are located, the maximum number of floors the building in which properties are located have, the price of properties and the longitude and latitude of properties.

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# Loading the dataset into a pandas dataframe
df1 = pd.read_csv('uybor_clean.csv')
```

## Distribution of Real Estate Properties Among the Districts of Uzbekistan's Capital City.
The districts of Tashkent, provided in this dataset, accomodate numerous amounts of real estate properties up for sale. The Chilonzor district has the most real estate properties up for sale with over 1600 properties, while the Bektemir district has the least amount of properties up for sale with a count of just 8 properties.

```python
# setting a base color to blue

base_color = sns.color_palette()[0]

# plotting a bar chart to show the frequencies of apartments availabele for sale in the different districts.

type_order = df1.district.value_counts().index
sns.countplot(data=df1, x='district', color=base_color, order=type_order)
plt.title('Distribution of Properties Across Districts')
plt.ylabel('Frequency')
plt.xlabel('Districts')
plt.xticks(rotation=90);
```

## Number of Rooms Real Estate Properties Are Built With. 
7 out of the 12 districts present in this dataset, mostly have 2 room properties up for sale in them with the rest having 2 room properties as the second most popular properties up for sale. It shows that people who generally own properties and live in Tashkent do not have large families.

```python
def countplot(data, x, y):
    sns.countplot(data=data, x=x, hue=y)

# plotting a bar chart showing the most common number of rooms most properties in the districts of Tashkent come with.
plt.figure(figsize=[10,10])
plt.subplot(2,1,1)
type_order = df1.rooms.value_counts().index
sns.countplot(data=df1, x='rooms', color=base_color, order=type_order)
plt.title('Distribution Of The Number Of Rooms Properties Contain')
plt.ylabel('Frequency')
plt.xlabel('Rooms');
    
# showing relationships between the various districts in Tashkent state and the number of rooms that the properties in these
#  these districts come with.
plt.figure(figsize=[20,10])
countplot(df1, 'district', 'rooms')
plt.title('Distribution of Rooms In Properties Across the Districts')
plt.xlabel('District')
plt.ylabel('Frequency')
plt.xticks(rotation=90);
```

## Average Price of Real Estate Properties  Across Districts
Houses on average in Tashkent, generally cost below 100,000 USD. Also, the Mirobod district is glaringly the most expenisve district on average, with the Bektemir district being the district that boasts the cheapest real estate properties up for sale on the market.

```python
def boxplot(data, x, y):
    sns.boxplot(data=data, x=x, y=y, color=base_color);

# making a dataframe with reduced price caps
df_cut = df1[df1.price_in_usd <= 180000]

# plotting a box plot to show the relationship between the location and price of real estate in the districts of tashkent
plt.figure(figsize=[10,8])
boxplot(data=df_cut, x='district', y='price_in_usd');
plt.title('Relating Properties in Districts With Their Prices')
plt.xlabel('Districts')
plt.ylabel('Price ($)')
plt.xticks(rotation=90);
```

## Price by District and Room Size of Property
Real estate properties tend to see an increase in price as the number of rooms they possess increases across the districts in this dataset. This follows societal norms as bigger roomed properties are ideally expensive than smaller ones in same areas.

```python
# using a clustered bar chart to see how the number of rooms a property has influences its price.
plt.figure(figsize=[20,20])
ax = sns.barplot(data=df1, x='district', y='price_in_usd', hue='rooms')
ax.legend(framealpha=1, title='rooms')
plt.title('How The Number of Rooms of Properties Across Districts Influences Its Price')
plt.xlabel('District')
plt.ylabel('Price ($)')
plt.xticks(rotation=90);
```

## Price by Size of Properties across Districts
Scatterplots relating the price of real estate properties to their sizes show a positive correlation, therefore showing that an increase in the size of real estate properties across the various districts in this dataset brings about an increase in the price of said properties.

```python
# adapting bivariate plots to show multivariate relationships
# visualizing how the size of properties influences the price of properties in the various districts in this dataset 
g = sns.FacetGrid(data=df1, col='district', col_wrap=4)
g.map(plt.scatter, 'size_(m2)', 'price_in_usd');
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('How The Size of Properties Relates With Its Price Across The Districts');
```
